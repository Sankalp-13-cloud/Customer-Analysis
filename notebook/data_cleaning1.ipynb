```python
import pandas as pd
```

```python
# Load the data
df = pd.read_csv("d:/Datamites/Projects/cus_seg/data/raw/rawfile1.csv")
```

```python
# Checking for null values
df.isnull().sum()
```

```python
# Checking data type of each attribute
df.info()
```

```python
# Converting invoice_date from object to datetime
df['invoice_date'] = pd.to_datetime(df['invoice_date'], format='%d-%m-%Y')

print(df.dtypes)
```

```python
df.head()
```

```python
# Checking for duplicates
df.duplicated().sum()
```

```python
# Clean column names properly
df.columns = df.columns.str.strip()               # remove extra spaces
df.columns = df.columns.str.title()                # make all sentance case
df.columns = df.columns.str.replace(' ', '_')      # replace spaces with underscores

df.head()
```

```python
# Clean text data inside cells
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].str.strip()

df.head()
```

```python
# Specify the file path for the processed file
output_file_path = r"D:/Datamites/Projects/cus_seg/data/processed/processedfile1.csv"

# Export the cleaned dataframe to CSV
df.to_csv(output_file_path, index=False)  # index=False to avoid saving row indices
```
